 Merging CSV and JSON data - Products table

```python
# Merge products dataframes on 'Uniq_id'
merged_products = pd.concat([products_csv.set_index('Uniq_id'), products_json.set_index('uniq_id')], axis=1, join='inner').reset_index()
merged_products.head(2)
```

 Merging CSV and JSON data - Products,reviews table

```python
#Breaking the reviewed column of reviews to find the matching uniq id with products table
products_exploded = reviewers_json.explode('Reviewed')
```

```python
reviews_products_list = pd.merge(merged_products, products_exploded, left_on='index', right_on='Reviewed', how='inner')
```

##  Missing values altering

```python
#As the space value is not recognized as a null value, it is modified to null to identify the original empty values
merged_products.replace('', pd.NA, inplace=True)
merged_products.isnull().sum()
```

```python
#Updating the category with most repeated category name
mode=merged_products['category'].mode()
merged_products['category'] = merged_products['category'].fillna(mode)
```

```python
# Dropping the null values of sale price with its mean
merged_products['sale_price'] = pd.to_numeric(merged_products['sale_price'], errors='coerce')

mean=merged_products['sale_price'].mean()
merged_products['sale_price'] = merged_products['sale_price'].fillna(mean)
```

### **Analysing the jc penny dataset**

**1. Sentiment analysis for reviews for top 10 brands**



```python
# Function to extract the  review from reviews column
def extract_first_review(reviews):
    if isinstance(reviews, list) and len(reviews) > 0:
        return reviews[0].get('Review')
    return None

# Applying the function to extract the first review and creating a new 'Review' column
reviews_products_list['Review'] = reviews_products_list['Reviews'].apply(extract_first_review)

```

```python
from textblob import TextBlob

# Function to determine sentiment to check if it postive,negative or neutral comment
def analyze_sentiment(review):
    if review:
        analysis = TextBlob(review)
        # Classify as positive, negative, or neutral
        return 'Positive' if analysis.sentiment.polarity > 0 else 'Negative' if analysis.sentiment.polarity < 0 else 'Neutral'
    return None

# Applying sentiment analysis to the 'Review' column and creating a new 'Sentiment' column
reviews_products_list['Sentiment'] = reviews_products_list['Review'].apply(analyze_sentiment)

# Displaying the resulting DataFrame with 'Sentiment' column
reviews_products_list[['brand', 'Review', 'Sentiment']]

```

THis graph shows that most of reviews are positive for all the brands

```python
import plotly.express as px

# Extracting Top 10 brands
top_brands = reviews_products_list['brand'].value_counts().nlargest(10).index

# Filter  for top 10 brands
top_brands_df = reviews_products_list[reviews_products_list['brand'].isin(top_brands)]

# Grouping by brand and sentiment.Then calculate its proportions
sentiment_proportions = (
    top_brands_df.groupby(['brand', 'Sentiment'])
    .size()
    .unstack(fill_value=0)
    .apply(lambda x: x / x.sum(), axis=1)
    .stack()
    .reset_index(name='Proportion')
)

# Creating a bar chart using Plotly Express
fig = px.bar(sentiment_proportions, x='brand', y='Proportion', color='Sentiment',
             labels={'Proportion': 'Sentiment Proportion'}, barmode='group',
             title='Sentiment Proportion for Top 10 Brands')

# Showing the figure
fig.show()
```

2. Age group distribution

From the below graph we can confirm that most reviewed people are falling in the age group between 61 to 70 and least reviwed age group is 71+

```python
# Converting 'DOB' to datetime
from datetime import datetime
reviews_products_list['DOB'] = pd.to_datetime(reviews_products_list['DOB'])

# Calculating age
reviews_products_list['age'] = (datetime.now() - reviews_products_list['DOB']).astype('<m8[Y]')

# Defining age groups
bins = [0, 18, 30, 40, 50, 60, 70, 120]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+']

# Creating 'Age Group' column
reviews_products_list['Age Group'] = pd.cut(reviews_products_list['age'], bins=bins, labels=labels, right=False)

# Counting the occurrences of each age group
age_group_counts = reviews_products_list['Age Group'].value_counts()

# Creating a pie chart
fig = px.pie(age_group_counts, names=age_group_counts.index, values=age_group_counts.values,
             title='Proportion of Age Groups',
             labels={'index': 'Age Group', 'value': 'Count'},
             hole=0.4)

# Showing the figure
fig.show()
```
